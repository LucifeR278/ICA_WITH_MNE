# Wiener Filtering

image.png


```python
import numpy as np
import mne
from scipy.signal import wiener

# Load raw EEG data
#raw = mne.io.read_raw_fif('file_raw.fif', preload=True)
sample_data_folder = mne.datasets.sample.data_path()
sample_data_raw_file = sample_data_folder / "MEG" / "sample" / "sample_audvis_raw.fif"
raw = mne.io.read_raw_fif(sample_data_raw_file, verbose=False,preload=True).crop(tmax=60)

# Extract the data and sampling frequency
data = raw.get_data()
sfreq = raw.info['sfreq']

# Apply Wiener filter to each channel
filtered_data = np.zeros_like(data)
for i in range(data.shape[0]):
    filtered_data[i] = wiener(data[i])

# Create a new Raw object with the filtered data
raw_filtered = mne.io.RawArray(filtered_data, raw.info)

# Plot the original and Wiener filtered signals for comparison
raw.plot(n_channels=10, title='Original Signal')
raw_filtered.plot(n_channels=10, title='Wiener Filtered Signal')
```

```python
import numpy as np
import mne

# Load raw EEG data
#raw = mne.io.read_raw_fif('file_raw.fif', preload=True)
sample_data_folder = mne.datasets.sample.data_path()
sample_data_raw_file = sample_data_folder / "MEG" / "sample" / "sample_audvis_raw.fif"
raw = mne.io.read_raw_fif(sample_data_raw_file, verbose=False,preload=True).crop(tmax=60)


# Extract data and sampling frequency
data = raw.get_data()
sfreq = raw.info['sfreq']

# Assume the signal of interest is within a certain frequency band, e.g., 8-12 Hz (alpha band)
l_freq = 8.0
h_freq = 12.0

# Band-pass filter the data to isolate the signal of interest
filtered_data = raw.copy().filter(l_freq=l_freq, h_freq=h_freq).get_data()

# Calculate the power of the filtered (signal) and unfiltered (signal + noise) data
signal_power = np.mean(filtered_data**2, axis=1)
total_power = np.mean(data**2, axis=1)

# Estimate noise power as the difference between total power and signal power
noise_power = total_power - signal_power

# Calculate SNR (in linear scale)
snr = signal_power / noise_power

# Convert SNR to decibels (dB)
snr_db = 10 * np.log10(snr)

# Print SNR for each channel
for ch, snr_value in zip(raw.ch_names, snr_db):
    print(f"Channel {ch}: SNR = {snr_value:.2f} dB")
```

: Higher SNR values (in dB) indicate better signal quality. Negative values can occur if the noise power is greater than the signal power.
